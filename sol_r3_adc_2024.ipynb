```python
from mlxtend.frequent_patterns import apriori, association_rules

data_apriori = data.copy()
```

```python
# 2a

data_apriori = data_apriori[films]
data_apriori = (data_apriori >= 0).astype(int)  
data_apriori.head()
```

```python
%%time

# Building the model
frq_items = apriori(data_apriori, min_support = 0.5, use_colnames = True)
```

```python
frq_items
```

```python
# Collecting the inferred rules in a dataframe 
rules = association_rules(frq_items, metric ="lift") 
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False]) 
print(rules.head())
```

```python
rules
```

```python
rules.sort_values(by=['support'], ascending=False)
```

```python
# top 10 lift

rules.nlargest(10, "lift")
```

```python
from apyori import apriori

records = []  
for i in range(0, len(data_apriori)):
    tmp2 = []
    for j in range(0,len(data_apriori.columns)):
        if data_apriori.values[i,j] == 1:
            tmp2.append(data_apriori.columns[j]) 
    records.append(tmp2)
```

```python
records = records[0:250]
len(records)
```

```python
%%time
association_rules = apriori(records, min_support=0.35, min_confidence=0.50, min_length=2)
association_results = list(association_rules)  
```

```python
association_results
```

```python
results = []
count = 0
for item in association_results:
    count += 1
    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    #print(pair, items, items[0])
    if len(items) != 1:
        value0 = str(items[0])
        value1 = str(items[1])

        #second index of the inner list
        value2 = str(item[1])[:7]

        #third index of the list located at 0th
        #of the third index of the inner list

        value3 = str(item[2][0][2])[:7]
        value4 = str(item[2][0][3])[:7]

        rows = (value0, value1,value2,value3,value4)
        results.append(rows)
    
labels = ['Title 1','Title 2','Support','Confidence','Lift']
movie_suggestion = pd.DataFrame.from_records(results, columns = labels)
```

```python
from surprise import SVD, Reader
from surprise import Dataset
from surprise.model_selection import cross_validate
```

```python
data_surprise = data[['ClientID', 'Gender', 'Age',
       'MarvelFavoriteCharacter', 'Iron Man',
       'The Incredible Hulk', 'Iron Man 2', 'Thor',
       'Captain America: The First Avenger', 'The Avengers', 'Iron Man 3',
       'Thor: The Dark World', 'Captain America: The Winter Soldier',
       'Guardians of Galaxy', 'Avengers: Age of Ultron', 'Ant-Man',
       'Captain America: Civil War', 'Doctor Strange', 'Guardians of Galaxy 2',
       'Spider-Man: Homecoming', 'Thor: Ragnarok', 'Black Panther',
       'Avengers: Infinity War', 'Ant-Man and the Wasp', 'Captain Marvel',
       'Avengers: Endgame', 'Spider-Man: Far From Home', 'Black Widow',
       'Shang-Chi and the Legend of the Ten Rings', 'Eternals',
       'Spider-Man: No Way Home' ]]
```

```python
# we have to change the dataset. We will make it with userid, movie, rating, and the features of each user.

data_surprise = pd.DataFrame(data_surprise.set_index(['ClientID', 'Gender', 'Age',
       'MarvelFavoriteCharacter']).stack())

data_surprise.columns = ['Rating']
data_surprise = data_surprise.reset_index()
data_surprise.columns = ['ClientID', 'Gender', 'Age', 'MarvelChar', 'MovieID','Rating']
print(len(data_surprise))
data_surprise = data_surprise[data_surprise["Rating"] != -1]
print(len(data_surprise))
data_surprise.head()
```

```python
reader = Reader(rating_scale=(0, 10))
data_surprise2 = Dataset.load_from_df(data_surprise[['ClientID', 'MovieID', 'Rating']], reader)
```

```python
%%time
svd = SVD(verbose=True, n_epochs=50)
cross_validate(svd, data_surprise2, measures=['RMSE', 'MAE'], cv=5)
```

```python
%%time
trainset = data_surprise2.build_full_trainset()
svd.fit(trainset)
```
